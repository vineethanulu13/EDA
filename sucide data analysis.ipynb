# heatmap of indian states and union teritories taking age group as a criterion

```python
plt.figure(figsize=(12,10))
plt.style.use('fivethirtyeight')
sns.set_context('notebook',font_scale=1)
sns.set_style('whitegrid')
st= data.pivot_table(values='Total',index='State',columns='Age_group')
sns.heatmap(st,linewidths=3,cmap= 'RdYlBu_r')
```

# heatmap of indian states and union teritories taking year as a criterion

```python
plt.figure(figsize=(12,10))
plt.style.use('fivethirtyeight')
sns.set_context('notebook',font_scale=1)
sns.set_style('whitegrid')
st= data.pivot_table(values='Total',index='State',columns='Year')
sns.heatmap(st,linewidths=3,cmap='Set3')
```
